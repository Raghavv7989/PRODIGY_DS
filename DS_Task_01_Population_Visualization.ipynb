```python

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

file_path = 'population_data.csv'  # Replace with your path
df = pd.read_csv('population_data.csv', skiprows=4)
df.head()
```

```python

df.columns.values[:5]  
```

```python

df_2022 = df[['Country Name', '2022']].dropna()
df_2022['2022'] = pd.to_numeric(df_2022['2022'], errors='coerce')
df_2022 = df_2022.sort_values('2022', ascending=False)
df_2022.head()
```

```python

top10 = df_2022.head(10)
plt.figure(figsize=(12, 6))
plt.bar(top10['Country Name'], top10['2022'], color='skyblue')
plt.title('Top 10 Most Populated Countries (2022)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.hist(df_2022['2022'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Histogram of Country Populations (2022)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```
